```python
import googleapiclient.discovery
import pandas as pd

api_service_name = "youtube"
api_version = "v3"


youtube = googleapiclient.discovery.build(
    api_service_name, api_version, developerKey=DEVELOPER_KEY)


def getcomments(video):
  request = youtube.commentThreads().list(
      part="snippet",
      videoId=video,
      maxResults=100
  )

  comments = []

  # Execute the request.
  response = request.execute()

  # Get the comments from the response.
  for item in response['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      public = item['snippet']['isPublic']
      comments.append([
          comment['authorDisplayName'],
          comment['publishedAt'],
          comment['likeCount'],
          comment['textOriginal'],
          comment['videoId'],
          public
      ])

  while (1 == 1):
    try:
     nextPageToken = response['nextPageToken']
    except KeyError:
     break
    nextPageToken = response['nextPageToken']
    # Create a new request object with the next page token.
    nextRequest = youtube.commentThreads().list(part="snippet", videoId=video, maxResults=100, pageToken=nextPageToken)
    # Execute the next request.
    response = nextRequest.execute()
    # Get the comments from the next response.
    for item in response['items']:
      comment = item['snippet']['topLevelComment']['snippet']
      public = item['snippet']['isPublic']
      comments.append([
          comment['authorDisplayName'],
          comment['publishedAt'],
          comment['likeCount'],
          comment['textOriginal'],
          comment['videoId'],
          public
      ])

  df2 = pd.DataFrame(comments, columns=['author', 'updated_at', 'like_count', 'text','video_id','public'])
  return df2
```

```python
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
from urllib.parse import urlparse
import whisper
import os

def get_transcript_from_url(url):
    try:
        # Method 1: Try using YouTubeTranscriptApi first (faster if available)
        url_data = urlparse(url)
        video_id = url_data.query[2:]  # Extract video ID from URL (for links like "youtube.com/watch?v=VIDEO_ID")
        if not video_id:  # For shortened links like "youtu.be/VIDEO_ID"
            video_id = url_data.path[1:]

        try:
            script = YouTubeTranscriptApi.get_transcript(video_id)
            transcript = " ".join([text["text"] for text in script if text["text"] != "[Music]"])
            return transcript
        except:
            # Fallback to Method 2: Download audio and transcribe if no captions available
            yt = YouTube(url)
            stream = yt.streams.get_by_itag(251)  # 251 = webm audio
            stream.download(filename="temp_audio.webm")

            # Transcribe using Whisper
            model = whisper.load_model("base")
            result = model.transcribe("temp_audio.webm")

            # Clean up
            if os.path.exists("temp_audio.webm"):
                os.remove("temp_audio.webm")

            return result["text"]

    except Exception as e:
        return f"Error: {str(e)}"

# Example usage:
youtube_url = "https://www.youtube.com/watch?v=0FtcHjI5lmw"
transcript = get_transcript_from_url(youtube_url)
print(transcript)
```

```python
import pandas as pd
import googleapiclient.discovery
from pytube import YouTube
from youtube_transcript_api import YouTubeTranscriptApi
from urllib.parse import urlparse
import whisper
import os
import torch
from transformers import BertTokenizer, BertModel
import torch.nn as nn
from tqdm import tqdm
import google.generativeai as genai




# Configuration
DEVELOPER_KEY =  "AIzaSyCxF9EO5Ddk7EKWeQOrdl_7xwyPafs98M0"  # Your YouTube API key
GOOGLE_API_KEY = 'AIzaSyCxCJoOU1A5JPDAwtmpt5nr-Q97jTqLNzg' # For Gemini
MODEL_PATH = '/content/drive/MyDrive/cyberbullying/bert_model.pth'
 # Path to your trained BERT model
TOKENIZER_DIR = '/content/drive/MyDrive/cyberbullying/tokenizer_dir'


# Initialize models
class BERTClassifier(nn.Module):
    def __init__(self, bert_model_name, num_classes):
        super(BERTClassifier, self).__init__()
        self.bert = BertModel.from_pretrained(bert_model_name)
        self.dropout = nn.Dropout(0.1)
        self.fc = nn.Linear(self.bert.config.hidden_size, num_classes)

    def forward(self, input_ids, attention_mask):
        outputs = self.bert(input_ids, attention_mask=attention_mask)
        pooled_output = outputs.pooler_output
        x = self.dropout(pooled_output)
        logits = self.fc(x)
        return logits

# Load BERT model
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = BERTClassifier(bert_model_name='bert-base-uncased', num_classes=6)
model.load_state_dict(torch.load(MODEL_PATH, map_location=device))
model.eval()
tokenizer = BertTokenizer.from_pretrained(TOKENIZER_DIR)

# Configure Gemini
genai.configure(api_key=GOOGLE_API_KEY)
gemini_model = genai.GenerativeModel('gemini-1.5-flash')

# YouTube functions
def get_video_id(url):
    url_data = urlparse(url)
    video_id = url_data.query[2:]  # For links like "youtube.com/watch?v=VIDEO_ID"
    if not video_id:  # For shortened links like "youtu.be/VIDEO_ID"
        video_id = url_data.path[1:]
    return video_id

def get_transcript(video_id):
    try:
        # First try to get official captions
        script = YouTubeTranscriptApi.get_transcript(video_id)
        transcript = " ".join([text["text"] for text in script if text["text"] != "[Music]"])
        return transcript
    except:
        # Fallback to audio download + Whisper
        try:
            yt = YouTube(f"https://www.youtube.com/watch?v={video_id}")
            stream = yt.streams.get_by_itag(251)  # 251 = webm audio
            stream.download(filename="temp_audio.webm")

            whisper_model = whisper.load_model("base")
            result = whisper_model.transcribe("temp_audio.webm")

            if os.path.exists("temp_audio.webm"):
                os.remove("temp_audio.webm")

            return result["text"]
        except Exception as e:
            print(f"Error getting transcript: {str(e)}")
            return None

def get_comments(video_id, max_comments=20):
    youtube = googleapiclient.discovery.build(
        "youtube", "v3", developerKey=DEVELOPER_KEY)

    comments = []
    request = youtube.commentThreads().list(
        part="snippet",
        videoId=video_id,
        maxResults=min(100, max_comments),
        order="relevance"  # Get top comments first
    )

    try:
        response = request.execute()
        for item in response['items']:
            if len(comments) >= max_comments:
                break
            comment = item['snippet']['topLevelComment']['snippet']
            comments.append({
                'author': comment['authorDisplayName'],
                'text': comment['textOriginal'],
                'likes': comment['likeCount'],
                'published_at': comment['publishedAt']
            })
    except Exception as e:
        print(f"Error getting comments: {str(e)}")

    return comments[:max_comments]

# Analysis functions
def predict_cyberbullying(text):
    classes = ['age', 'ethnicity', 'gender', 'not_cyberbullying', 'other_cyberbullying', 'religion']

    inputs = tokenizer(text, return_tensors='pt', max_length=128, padding='max_length', truncation=True)
    input_ids = inputs['input_ids'].to(device)
    attention_mask = inputs['attention_mask'].to(device)

    with torch.no_grad():
        outputs = model(input_ids=input_ids, attention_mask=attention_mask)
        _, preds = torch.max(outputs, dim=1)

    return {
        'prediction': classes[preds.item()],
        'is_cyberbullying': classes[preds.item()] != 'not_cyberbullying'
    }

def transform_negative_text(text):
    prompt = f"""You are a text transformation assistant that specializes in converting negative or cyberbullying content into positive, constructive messages while preserving the original intent where possible.

        Instructions:
        1. Identify any negative language, insults, cyberbullying, or harmful content
        2. Remove or replace problematic words/phrases
        3. Rephrase the message to be positive and constructive
        4. Keep the core meaning where appropriate
        5. If the text is already positive, return it unchanged
        6. DOnt add extra your comments just give the paarphrased content .
    Original: {text}

    Transformed: """

    try:
        response = gemini_model.generate_content(prompt)
        return response.text if response.text else text
    except:
        return text

# Main analysis function
def analyze_youtube_video(url):
    video_id = get_video_id(url)
    if not video_id:
        return None

    # Get content
    transcript = get_transcript(video_id)
    comments = get_comments(video_id)

    # Prepare results DataFrame
    results = []

    # Analyze transcript in chunks if too long
    if transcript:
        chunk_size = 500  # characters per chunk
        transcript_chunks = [transcript[i:i+chunk_size] for i in range(0, len(transcript), chunk_size)]

        for i, chunk in enumerate(transcript_chunks):
            analysis = predict_cyberbullying(chunk)
            transformed = transform_negative_text(chunk) if analysis['is_cyberbullying'] else chunk

            results.append({
                'type': 'transcript',
                'chunk_num': i+1,
                'original_text': chunk,
                'prediction': analysis['prediction'],
                'is_cyberbullying': analysis['is_cyberbullying'],
                'transformed_text': transformed
            })

    # Analyze comments
    for comment in comments:
        analysis = predict_cyberbullying(comment['text'])
        transformed = transform_negative_text(comment['text']) if analysis['is_cyberbullying'] else comment['text']

        results.append({
            'type': 'comment',
            'author': comment['author'],
            'original_text': comment['text'],
            'prediction': analysis['prediction'],
            'is_cyberbullying': analysis['is_cyberbullying'],
            'transformed_text': transformed,
            'likes': comment['likes'],
            'published_at': comment['published_at']
        })

    return pd.DataFrame(results)

# Example usage

```
